```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
np.random.seed(1)
tf.random.set_seed(346)

# load json and create model
json_file = open('model_smallGPUGPU.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = keras.models.model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model_smallGPUGPU.h5")

# evaluate loaded model on test data
opt = keras.optimizers.SGD(learning_rate=0.01)
loaded_model.compile(optimizer=opt, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

for ii in range(11):
    
    print('Start with %d'%ii+'\n')
    
    train = np.load('../data/data_split/train%d'%ii+'.npy') 
    test = np.load('../data/data_split/test%d'%ii+'.npy') 

    train_targ = np.load('../data/data_split/train_targ_%d'%ii+'.npy') 
    test_targ = np.load('../data/data_split/test_targ_%d'%ii+'.npy') 

    print(len(train),len(train_targ))
    print(len(test),len(test_targ))

    feat_tr2 = train.reshape(len(train), 51, 153, 1)
    feat_te2 = test.reshape(len(test), 51, 153, 1)

    # -- print the accuracy
    loss_tr, acc_tr = loaded_model.evaluate(feat_tr2, train_targ,verbose=False)
    loss_te, acc_te = loaded_model.evaluate(feat_te2, test_targ,verbose=False)

    print("Training accuracy : {0:.4f}".format(acc_tr))
    print("Testing accuracy  : {0:.4f}".format(acc_te))

#     file1 = open("results.txt","a")
#     file1.write("Training accuracy correct GPU for stamps%d"%ii+": {0:.4f} \n".format(acc_tr))
#     file1.write("Testing accuracy correct GPU for stamps%d"%ii+": {0:.4f} \n".format(acc_te))
#     file1.close()

    train = None
    test = None
    train_targ = None
    test_targ = None
```
